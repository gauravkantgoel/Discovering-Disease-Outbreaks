```python
import geonamescache
import unidecode
import re
```

```python
# Instatntiate an object from the amazing geonamescache library

gc = geonamescache.GeonamesCache()
```

```python
# Countries Names

gc.get_countries_by_names().keys()
```

```python
countries = []

for key in gc.get_countries_by_names().keys():
    countries.append(key)
```

```python
cities = []

for city in gc.get_cities().values():
    cities.append(city['name'])
```

```python
def name_to_regex(name):
    decoded_name = unidecode.unidecode(name)
    if name != decoded_name:
        regex = fr'\b({name}|{decoded_name})\b'
    else:
        regex = fr'\b{name}\b'
    return re.compile(regex, flags=re.IGNORECASE)
```

```python
country_to_name = {}
for country in countries:
    country_to_name[name_to_regex(country)]= country
```

```python
country_to_name
```

```python
city_to_name = {}
for city in cities:
    city_to_name[name_to_regex(city)]= city
```

```python
 city_to_name
```

```python
def get_match(text, dictionary):
    for regex, name in sorted(dictionary.items(),
                              key=lambda x: x[1]):
        if regex.search(text):
            return name
    return None
```

```python
import pandas as pd

matched_countries = [get_match(headline, country_to_name)
                     for headline in headlines]
matched_cities = [get_match(headline, city_to_name)
                  for headline in headlines]
data = {'Headline': headlines, 'City': matched_cities,
        'Country': matched_countries}
df = pd.DataFrame(data)
```
